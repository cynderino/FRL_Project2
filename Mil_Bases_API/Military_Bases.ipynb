```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
from sqlalchemy import create_engine

# PyMySQL 
import pymysql
pymysql.install_as_MySQLdb()

import numpy as np
import requests
import time
from datetime import datetime
from datetime import timedelta
from citipy import citipy

from config import mypass
```

```python
query_url = "https://public.opendatasoft.com/api/records/1.0/download/?dataset=military-bases&format=json&pretty_print=true"
state_list = []
site_name_list = []
status_list = []
component_list = []
latitude_list = []
longitude_list = []
country_list = []


# try:
#     mil_base_response = requests.get(query_url)
#     mil_base_json = mil_base_response.json()

#     for i in np.arange(len(mil_base_json)):                

#             state_list.append(mil_base_json[i]['fields']['state_terr'])
#             site_name_list.append(mil_base_json[i]['fields']['site_name'])
#             status_list.append(mil_base_json[i]['fields']['oper_stat'])
#             component_list.append(mil_base_json[i]['fields']['component'])
#             latitude_list.append(mil_base_json[i]['fields']['geo_point_2d'][0])
#             longitude_list.append(mil_base_json[i]['fields']['geo_point_2d'][1])
#             country_list.append(mil_base_json[i]['fields']['country'])
    
        
# except Exception as e:
#     print(e)
#     print(query_url)

try:
    mil_base_response = requests.get(query_url)
    mil_base_json = mil_base_response.json()

    for i in np.arange(len(mil_base_json)):                

            state_list.append(mil_base_json[i]['fields']['state_terr'])
            site_name_list.append(mil_base_json[i]['fields']['site_name'])
            status_list.append(mil_base_json[i]['fields']['oper_stat'])
            component_list.append(mil_base_json[i]['fields']['component'])
            country_list.append(mil_base_json[i]['fields']['country'])
            
            try:
                latitude_list.append(mil_base_json[i]['fields']['geo_point_2d'][0])
                longitude_list.append(mil_base_json[i]['fields']['geo_point_2d'][1])
            
            except Exception as e:
                latitude_list.append(0)
                longitude_list.append(0)

except Exception as e:
    print(e)
    print(query_url)

```

```python
len(latitude_list) 
```

```python
mil_base_dict = {
    "State": state_list,
    "Site_Name": site_name_list,
    "Operational_Status": status_list,
    "Component": component_list,
    "Latitude": latitude_list,
    "Longitude": longitude_list,
    "Country": country_list,
}
        
mil_base_df = pd.DataFrame(mil_base_dict)
mil_base_df.count()
```

```python
cleaned_mil_base_df = mil_base_df[mil_base_df.Country == 'United States']
cleaned_mil_base_df.count()
```
