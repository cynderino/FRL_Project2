```python
stop = stopwords

# Exclude stopwords with Python's list comprehension and pandas.DataFrame.apply.
description_df['Scrub Description'] = description_df['Short Description'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
print(description_df)
```

```python
# Variable Top 25 Words
Top25 = Counter(" ".join(description_df['Scrub Description'].str.lower()).split()).most_common(25)

# Create Dataframe
word_cloud_df = pd.DataFrame(Top25)
word_cloud_df = word_cloud_df.rename(columns={0:'Word',1:'Count'})
word_cloud_df
```
